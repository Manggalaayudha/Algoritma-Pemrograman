Membuat Fungsi Dengan Argumen

```python
# def nama fungsi():
#   badan fungsi

# def hello_world(argumen, parameter, input):

def hello_world(nama):
  print(f'Selamat datang di dunia, {nama}')

hello_world("zeus")
hello_world("ucup")

def tambah(angka_1, angka_2):
  hasil = angka_1 + angka_2
  print(f'{angka_1} + {angka_2} = {hasil}')

tambah(34, 54)

def say_hi(list_peserta):
  data_peserta = list_peserta.copy()
  for peserta in data_peserta:
    print(f"Selamat Datang {peserta}")

kumpulan_mahasiswa = ["Joko", "Budi", "Sutono"]

say_hi(kumpulan_mahasiswa)
```

Latihan

```python
# membuat program dengan input dan fungsi untuk menghitung:
# a. luas persegi panjang
# b. volume persegi panjang / balok
# c. luas lingkaran
# d. volume bola

import math

# a. Luas persegi panjang
def hitung_luas(panjang, lebar):
  return panjang * lebar
p = int(input("Masukkan Panjang : "))
l = int(input("Masukkan Lebar : "))

print(f"Luas Persegi Panjang = {hitung_luas(p, l)} \n ")

# b. Volume persegi panjang (balok)
def vol_bal(panjang, lebar, tinggi):
  return panjang * lebar * tinggi
p = int(input("Masukkan Panjang : "))
l = int(input("Masukkan Lebar : "))
t = int(input("Masukkan Tinggi : "))

print(f"Volume Balok = {vol_bal(p, l, t)} \n")

# c. Luas lingkaran
def luas_lingkaran(jari_jari):
    return math.pi * jari_jari ** 2
r = float(input("Masukkan jari-jari lingkaran: "))
print(f"Luas Lingkaran = {luas_lingkaran(r)} \n")

# d. Volume bola
def volume_bola(jari_jari):
    return (4/3) * math.pi * jari_jari ** 3

r = float(input("Masukkan jari-jari: "))

hasil = volume_bola(r)
print("Volume bola =", hasil)
```
