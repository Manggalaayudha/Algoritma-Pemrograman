```python
# Buatlah program sistem manajemen perpustakaan sederhana (menambahkan dan
# menampilkan buku baru ke dalam sistem) menggunakan konsep multi-keys dan nesting
# dictionary dengan format
# - KEYS
# - JUDUL BUKU
# - PENULIS
# - TAHUN TERBIT
# - TANGGAL TERAKHIR PINJAM
# - STATUS (TERSEDIA / TIDAK)

import datetime
import string
import random

perpus_template = {
    'judul_buku' : 'judul',
    'penulis' : 'penulis',
    'tahun_terbit' : 0000,
    'terakhir_pinjam' : datetime.datetime(1111,11,11),
    'status' : True
}

data_perpus = {}
while True:
  print(f"{'SELAMAT DATANG':^20}")
  print(f"{'DAFTAR BUKU':^20}")
  print("-"*30)
  perpus = dict.fromkeys(perpus_template.keys())

  perpus['judul_buku'] = input("Judul Buku : ")
  perpus['penulis'] = input("Penulis : ")
  perpus['tahun_terbit'] = input("Tahun Terbit : ")
  TAHUN_PINJAM = int(input ("Tahun Pinjam (YYYY) : "))
  BULAN_PINJAM = int(input ("Bulan Pinjam (1-12) : "))
  TANGGAL_PINJAM = int(input ("Tanggal Pinjam (1-31) : "))
  perpus['status'] = input("Status (tersedia/tidak) : ")
  perpus['terakhir_pinjam'] = datetime.datetime(TAHUN_PINJAM, BULAN_PINJAM, TANGGAL_PINJAM)

  KEY = ''.join((random.choice(string.ascii_uppercase) for i in range (6)))
  data_perpus.update({KEY:perpus})
  print(f"{'KEY':<9} {'JUDUL' : <10} {'PENULIS' : <9} {'TERBITAN' : <11} {'DIPINJAM' : <11} {'STATUS' : <10}")
  print('-'*70)

  for perpus in data_perpus:
    KEY = perpus

    JUDUL = data_perpus [KEY] ['judul_buku']
    PENULIS = data_perpus [KEY] ['penulis']
    TERBIT = data_perpus [KEY] ['tahun_terbit']
    PINJAM = data_perpus [KEY] ['terakhir_pinjam'].strftime("%x")
    STATUS = data_perpus [KEY] ['status']

    print(f"{KEY:<9} {JUDUL: <10} {PENULIS : <11} {TERBIT : <9} {PINJAM : <11} {STATUS : <10}")
  print("\n")
  is_done = input("Apakah ada data yang ingin ditambahkan? (y/n) : ")
  if is_done == "n" :
    break
```
