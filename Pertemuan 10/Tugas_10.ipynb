```python
# Buat directory keranjang dengan data: {"apel" : 5000, "pisang" : 3000, "jeruk" : 7000, "anggur" : 15000}
# gunakan perulangan untuk menghitung total harga semua barang
# tampilkan setiap barang dan harganya dalam format : ex: "Harga apl : Rp 5000"
# tampilkan harga total belanja

keranjang = {
    "apel" : 5000,
    "pisang" : 3000,
    "jeruk" : 7000,
    "anggur" : 15000
}


total = 0

for buah, harga in keranjang.items():
  print(f"Harga buah {buah} : Rp{harga}")
  total = total + harga
print(f"\nharga total : {total}")
```
