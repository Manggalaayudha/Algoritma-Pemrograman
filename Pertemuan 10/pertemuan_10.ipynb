Dictionary

```python
## Dictionary

# List -> array, diakses menggunakan index

data_list = ['ucup','salsa','xinthink']
print(data_list[2])

# Dictionary (dict) --> array asosiatif
# menggunakan identifier dan key

# data_dict = {
#     'key':'value',
# }

data_dict = {
    'cp':'ucup',
    'sl':'salsa',
    'xin':'xin think',
    'nmb': 100,
    'list': data_list
}
print(data_dict) #untuk memanggil semua
print(data_dict['sl']) #untuk memanggil salah satu data
print(data_dict['nmb'])
```

Operasi Dictionary

```python
## Operasi Dictionary

data_dict = {
  "cup" : "ucup surucup",
  "pet" : "opet saripet",
  "ndo" : "nando surando"
}

# # melihat panjang data dictionary
dict_len = len(data_dict)
print(f"panjang dictionary : {dict_len}")

# mengecek key eksis atau tidak
key = 'cup'
check_key = key in data_dict
print(f"apakah {key} ada di data_dict : {check_key}")

key1 = 'tong'
check_key = key1 in data_dict
print(f"apakah {key1} ada di data_dict : {check_key}")

# mengakses value (read) dengan get
print(data_dict["cup"])
print(data_dict.get("cup"))
print(data_dict.get("cup", "key tersebut tidak ditemukan"))
print(data_dict.get("tong", "key tersebut tidak ditemukan"))
print(data_dict.get("tong"))

# # mengupdate / mengubah data
print(f"sebelum diubah {data_dict}")

data_dict["cup"] = "ucup si ganteng"
print(f"setelah diubah {data_dict}")

print(f"sebelum ditambah : {data_dict}")
data_dict["de"] = "ade si lucu"
print(f"setelah ditambah : {data_dict}")

data_dict.update({"pet" : "opet si buhlil"})
data_dict.update({"ren" : "ren si oren"})
print(f"setelah ditambah lagi : {data_dict}")

# # menghapus data
print(f"sebelum dihapus {data_dict}")

del data_dict["cup"]
print(f"setelah dihapus {data_dict}")
```

Looping Dictionary

```python
teman_teman = {
    'cup'   : 'ucup surucup',
    'tong'  : 'sotong surotong',
    'lur'   : 'lur sedulur',
    'wel'   : 'welll bung towellll'
}

print(teman_teman)

# agar lebih rapi bisa menggunakan perulangan (looping)

for teman in teman_teman:
  print(teman)

# operator untuk mengambil item
keys = teman_teman.keys()
print(keys)

for key in teman_teman.keys():
  print(key)

values = teman_teman.values()
print(values)

for value in teman_teman.values():
  print(value)

items = teman_teman.items()
print(items)

for item in teman_teman.items():
  print(item)

for key,value in teman_teman.items():
  print(f"key : {key},\nvalue nya adalah {value}\n")
```
